```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
df = pd.read_csv("./data/50_Startups.csv")

df.head()
```

```python
X = df[["R&D_Spend_KES", "Administration_KES", "Marketing_Spend_KES", "County"]].values  
#.values is converting dataframe into array
print(X)
y = df[["Profit_KES"]].values
print(y)
```

```python
st = ColumnTransformer([("County", OneHotEncoder(), [3])], remainder = 'passthrough')
X = st.fit_transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#Comparison

print('- y_pred : ')
print(y_pred)
print('- y_test : ')
print(y_test)
```

```python
print(y_pred.sum().mean())
print(y_test.sum().mean())

print("So called Accuracy ", y_pred.sum().mean()/y_test.sum().mean())
```
